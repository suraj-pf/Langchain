**LangChain Core Components: A Structural Overview**

**1. Models: The Foundation**

- **Purpose:**Â LangChain provides a standardized interface to interact with various AI models, preventing vendor lock-in.
    
- **Core Types:**
    
    - **LLMs (Large Language Models):**

    - **Chat Models:**
        
        - **Input/Output:**Â List of structured messages -> AI Message.

    - **Embedding Models:**
        
        - **Input/Output:**Â Text -> Vector (a list of numbers).
            
        - **Use Case:**Â Powering semantic search, clustering, and retrieval by converting text into numerical representations.
            

**2. Prompts: Engineering the Instructions**

- **Purpose:**Â To create dynamic, reusable, and effective instructions for the LLMs.
    
- **Key Techniques:**
    
    - **Prompt Templates:**Â Create prompts with placeholders (`{variable}`) to inject dynamic context, making prompts reusable.
        
    - **Role-Based Prompting:**Â UsingÂ `SystemMessage`Â to set the AI's behavior, personality, and rules at the start of a conversation.
        
    - **Few-Shot Prompting:**Â Including examples within the prompt to teach the model the desired format or style of response.
        

**3. Chains: Orchestrating Workflows**

- **Purpose:**Â To combine multiple components (models, prompts, logic) into a single, executable pipeline.
    
- **Core Concept:**Â The output of one step becomes the input to the next, automating multi-step processes.
    
- **Chain Types:**
    
    - **Sequential Chains:**Â A linear flow (e.g.,Â `Input -> LLM (Translate) -> LLM (Summarize) -> Output`).
        
    - **Parallel Chains:**Â Running multiple chains simultaneously on the same input and then combining the results (e.g., generating different sections of a report in parallel).
        
    - **Conditional Chains:**Â IncorporatingÂ `if/else`Â logic to create dynamic workflows based on the output of a previous step (e.g., routing user feedback to different endpoints).
        

**4. Indexes: Connecting to External Knowledge**

- **Purpose:**Â To give your LLM application access to private or specialized data it wasn't trained on (e.g., company PDFs, websites, databases). This is the foundation ofÂ **Retrieval-Augmented Generation (RAG)**.
    
- **The RAG Pipeline:**
    
    1. **Document Loader:**Â Loads data from a source (PDF, website, DB).
        
    2. **Text Splitter:**Â Breaks documents into smaller, semantically meaningful chunks.
        
    3. **Vector Store:**Â Creates embeddings for each chunk and stores them in a database.
        
    4. **Retriever:**Â Given a user query, it finds the most semantically relevant document chunks from the vector store and passes them to the LLM as context.
        

**5. Memory: Maintaining State**

- **Purpose:**Â To persist state/information between interactions with the LLM, as LLM API calls are inherently stateless.
    
- **Common Memory Types:**
    
    - **Conversation Buffer Memory:**Â Stores the raw text of the entire conversation.
        
    - **Conversation Buffer Window Memory:**Â Stores only the lastÂ `K`Â interactions to manage context length.
        
    - **Conversation Summary Memory:**Â Compresses the past conversation into a summary to save tokens.
        
    - **Custom Memory:**Â For storing specialized, structured data (e.g., user preferences, entity facts).
        

**6. Agents: AI with Reasoning and Tools**

- **Purpose:**Â To create "goal-oriented" AI that can reason and use tools to take actions.
    
- **Core Analogy:**Â An agent is a chatbot with superpowers.
    
    - **Chatbot:**Â Answers questions based on its internal knowledge.
        
    - **AI Agent:**Â _Reasons_Â about a goal,Â _plans_Â steps, andÂ _uses tools_Â (like a calculator, web search, or booking API) to achieve it.
        
- **Key Components:**Â An LLM for reasoning, a set ofÂ `Tools`Â it can use, and an agent executor that manages the decision-loop.
